```python
import pandas as pd
```

```python
GDP = pd.read_excel(r'C:\Users\Administrator\Desktop\API_NY.GDP.MKTP.CD_DS2_en_excel_v2.xls',skiprows = 3)
```

```python
GDP = GDP[['Country Name','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016']]
GDP.head()
```

```python
Top5 = ['United States','China','Japan','Germany','United Kingdom']
```

```python
GDP_Top5 = GDP[GDP['Country Name'].isin(Top5)].set_index('Country Name')
GDP_Top5
```

```python
GDP_Top5 = GDP_Top5.T
```

```python
GDP_Top5
```

```python
import matplotlib.pyplot as plt
```

```python
GDP_Top5.plot(figsize=(8,6))
plt.legend(loc= 'upper left',bbox_to_anchor=(1.0,1.0), frameon=False)
#plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')
plt.show()
```

```python
GDP_Top5.plot(kind = 'bar', figsize=(18, 10), rot = 0)
plt.legend(loc= 'upper left',bbox_to_anchor=(1.0,1.0), frameon=False)
plt.show()
```

```python
bars = GDP_Top5.plot(kind = 'barh', figsize=(10, 12), linewidth=0)
plt.legend(loc= 'upper left',bbox_to_anchor=(1.0,1.0), frameon=False)
plt.xlabel('GDP (current US$)', alpha=0.8, fontsize = 15)
plt.ylabel('Year', alpha=0.8, fontsize = 15)
for spine in plt.gca().spines.values():
    spine.set_visible(False)
# for bar in bars:
#     plt.gca().text(bar.get_x() + bar.get_width()/2, bar.get_height() - 5, str(int(bar.get_height())) + '%', 
#                  ha='center', color='w', fontsize=11)
plt.show()
```

```python
GDP_Top5.T.plot(kind = 'pie',subplots = True,layout = (5,2), fontsize = 15, legend = False, figsize=(15, 42), colormap='viridis')
plt.legend(loc= 'upper left',bbox_to_anchor=(1.0,5.8), frameon=False ,fontsize = 15)
plt.show()
```
